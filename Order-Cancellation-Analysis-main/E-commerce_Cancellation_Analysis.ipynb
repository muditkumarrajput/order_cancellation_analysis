```python
import pandas as pd
```

```python
df = pd.read_csv('E-commerce-data.csv')
```

```python
df
```

```python
df.shape
```

```python
df['Remarks'].value_counts()
```

```python
df_copy = df[df['Remarks'] != 'Verified Orders']
df_copy
```

```python
total_count = df_copy['Count'].sum()
total_count
```

```python
df_copy.groupby(['Remarks']).sum().plot(kind='pie', y='Count',autopct='%1.2f%%')
```

```python
df2 = df_copy.groupby(['Period','Remarks']).agg({'Count': ['sum']})
df2
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,7))
for name, group in df2.groupby('Remarks'):
    plt.plot(group['Period'],group['Count']['sum'], label=name)
plt.title('Line graph order Remarks by Period')
plt.xlabel("Period")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df3 = df_copy.groupby(['Channel', 'Remarks']).agg({'Count': ['sum']})
df3
```

df3.unstack().plot(kind='bar')
plt.title('Bar graph order Remarks by Channel')
plt.xlabel("Channel")
plt.ylabel("Count")
plt.legend()
plt.show()

```python
df4 = df_copy.groupby(['State']).agg({'Count': ['sum']})
df4
```

```python
df4.plot(kind='barh', figsize=(15,7))
plt.title('Bar graph order Remarks by State')
plt.xlabel("State")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df5 = df_copy.groupby(['Dest Zone', 'Remarks']).agg({'Count': ['sum']})
df5
```

```python
df5.unstack().plot(kind='bar', stacked=True)
plt.title('Bar graph order Remarks by Dest zone')
plt.xlabel("Dest zone")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df6 = df_copy.groupby(['Payment']).agg({'Count': ['sum']})
df6['% Share'] = 100 * df6[('Count', 'sum')] / df6[('Count', 'sum')].sum()
df6
```

```python
df6.plot(kind='pie', y='Count', autopct='%1.2f%%')
```
